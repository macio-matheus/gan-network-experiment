```python
import os

from keras.datasets import mnist, cifar100, fashion_mnist
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

from gan_network import GanNetwork
%matplotlib inline
```

```python
# Load dataset
```

```python
# Load the dataset
(x_train, _), (_, _) = mnist.load_data()
```

```python
path = "/home/jovyan/work/dataset/mnist/"

gan = GanNetwork()
gan.train(x_train, epochs=10, batch_size=64, sample_interval=100, path_save_image=path)
```

```python
# Exemple of image generated in 1000 epochs
paths = [os.path.join(path, file) for file in os.listdir(path)]
for p in paths:
    print(f"Epoch {(os.path.basename(p)).replace('.png', '')}")
    plt.imshow(mpimg.imread(p))
    plt.show()
```
